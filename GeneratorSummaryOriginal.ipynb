```python
path_save_data = "/content/drive/MyDrive/LLM_TEXT/inference/summary_original"
path_test_data = "/content/drive/MyDrive/LLM_TEXT/dataset/test_tokenized"
def write_data(offset, index, content):
    path = path_save_data +"/summary_{:06d}.txt".format(offset+index+1)
    print("Create file: ", path)
    f = open(path, "w")
    f.write(content)
    f.close()

def generatorSummary(offset, count):
  for i in range(count):
    print(i)
    path = path_test_data + "/{:06d}.txt.seg".format(i+offset+1)
    f = open(path, "r")
    lines = f.readlines()
    sum = ""
    content = ""
    empty = 0
    for line in lines:
      if(len(line) == 1):
        empty = empty +1
      elif(empty<1):
         sum += line + ". "
      elif(empty<2):
        sum +=line +" "
      elif(empty<3):
        content += line + " "
    sum = sum.replace("\n","")
    # sum = sum.replace("_"," ")
    # content = content.replace("\n","")
    # result = {
    #     'summary':sum,
    #     'content':content
    # }
    print ("sumary: ", sum)
    write_data(offset, i, sum)


```
