# **Dataset Evaluation**

```
# Đọc nội dung file
def read_content(path_file):
  ret = {}
  with open(path_file, "r", encoding="utf-8") as file:
      lines  = file.readlines()
      isTrue = False
      cnt = 0

      text_original = ""
      text_Summary = ""
      for line in lines:
          line = line.strip()
          if line == "":
            cnt += 1
            continue
          if cnt == 1:
            text_Summary += " " + line
          if cnt >=2:
            text_original += " " + line
            # print("", line)

  return {"text_summary": text_Summary, "text_original": text_original }
```

```
# Đọc danh sách các file từ đường dẫn được cung cấp
def listPaths(path):
  path += "train_tokenized" + '/*.txt.seg'
  pathfiles = list()
  cnt = 0
  for pathfile in glob.glob(path):
    cnt += 1
    pathfiles.append(pathfile)
  return pathfiles
```

```
def process_content_sentences( body: str, min_length=1, max_length=500):
    sents = [s.strip() for s in sent_tokenize(body)]
    return [c for c in sents if max_length > len(c.split(" ")) > min_length]
```

##**Dataset_3 Evaluation**

```
dataset_3_path = listPaths("/content/drive/MyDrive/LLM_TEXT/dataset/Datasets_3/")
for file in dataset_3_path:  print(file)
```

```
min_sentence_datasets3 = 1000
max_sentence_datasets3 = -1
min_word_number_datasets3 = 1000
max_word_number_datasets3 = -1
min_word_number_doc_datasets3 = 1000
max_word_number_doc_datasets3 = -1

for file in dataset_3_path:
  print(file)
  word_number_doc = 0
  content = read_content(file)
  text = content["text_original"]
  print("Content: ", text)
  sentences = process_content_sentences(text)
  sentence_number = len(sentences)
  print("sentencesentence number: ", sentence_number)
  if min_sentence_datasets3 > sentence_number:
    min_sentence_datasets3 = sentence_number

  if max_sentence_datasets3 < sentence_number:
    max_sentence_datasets3 = sentence_number

  for sentence in sentences:
    word_number = 0
    list_key = sentence.split(" ")
    for text_k  in  list_key:
      text_k = text_k.strip()
      if text_k != "":
        word_number += 1

    if min_word_number_datasets3 > word_number:
      min_word_number_datasets3 = word_number

    if max_word_number_datasets3 < word_number:
      max_word_number_datasets3 = word_number

    word_number_doc += word_number

  if min_word_number_doc_datasets3 > word_number_doc:
     min_word_number_doc_datasets3 = word_number_doc

  if max_word_number_doc_datasets3 < word_number_doc:
      max_word_number_doc_datasets3 = word_number_doc

```

```
print("Datasets3: evaluation with original content")
print("Minmum sentence/document is: ", min_sentence_datasets3)
print("Maxmum sentence/document is: ", max_sentence_datasets3)
print("Minmum word/sentence is: ", min_word_number_datasets3)
print("Maxmum word/sentence is: ", max_word_number_datasets3)
print("Minmum word/document is: ", min_word_number_doc_datasets3)
print("Maxmum word/document is: ", max_word_number_doc_datasets3)
```
